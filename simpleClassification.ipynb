```python
#Import required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn import linear_model
from sklearn.utils import shuffle
from sklearn.metrics import precision_score, confusion_matrix, recall_score, f1_score
import matplotlib.pyplot as plt
```

```python
#Store csv as pandas dataframe
data = pd.read_csv("Datasets/processedData.csv")
data = data.drop(data.columns[0],axis=1)
data.head(10)
```

```python
data = shuffle(data)
data.head(10)
```

```python
deptCode = LabelEncoder()
deptCode.fit(data['department'])
data['department'] = deptCode.transform(data['department'])

salaryCode = LabelEncoder()
salaryCode.fit(data['salary'])
data['salary'] = salaryCode.transform(data['salary'])
data.head(10)
```

```python
y = data.left
```

```python
def prediction_machine(user_def_test_size):
    x_train, x_test, y_train, y_test = train_test_split(data, y, test_size=user_def_test_size, random_state=2)
    clf = linear_model.SGDClassifier(max_iter=1000,tol=1e-3, random_state=2)
    clf.fit(x_train,y_train)
    model = cross_val_predict(clf, x_test, y_test, cv=5)
    print("Metrics with",(1-user_def_test_size),"% training data and ",(user_def_test_size),"% testing data")
    #Precision score
    y_true, y_prediction = y_test, model
    print("Precision score\t",precision_score(y_true, y_prediction))
    #Confusion matrix
    print("Confusion matrix")
    print(confusion_matrix(y_true, y_prediction))
    #Recall score
    print("Recall score\t",recall_score(y_true, y_prediction))
    #F1 score
    print("F1 score\t",f1_score(y_true, y_prediction))
    print()
```

```python
running = "yes"
while running == "yes":
    user_def_test_size = input("What percentage of the data do you want to use for testing? (Enter as a decimal between 0 and 1)")
    prediction_machine(float(user_def_test_size))
    running = input("Do you want to run the program again with a different test percentage? (Type 'yes' to run again)")
```
